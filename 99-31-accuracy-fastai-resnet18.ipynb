# Importing Libraries:

```python
from fastai.vision import *
from fastai.metrics import *
```

Below is the path to where my files are stored: 

```python
path = "../input/fruits/fruits-360"
```

Using the ImageDataBunch module I grab the data from the folders which is divided between training and test sets. The batch size is set at 16 for now, with image size being 360 x 360 px.

```python
data = ImageDataBunch.from_folder(path, ds_tfms = get_transforms(),
                                  train='Training', valid='Test',
                                  size=360, bs=16)
```

Lets visualize the data now:


```python
data.show_batch(4)
```

# Training and Predicting

The data looks crisp, lets try and classify it but first the model must be trained so lets train it on the training the data. I have decided to use the pretrained ResNet18 model:


```python
learn = cnn_learner(data, models.resnet18, metrics = accuracy, callback_fns = ShowGraph)
```

```python
learn.fit_one_cycle(2)
```

With just two epochs we literally reached almost perfect accuracy with out data.

Now its time to predict because without that we don't know if it actually does (I mean it does the metrics are supposed to be right).

```python
img = open_image("../input/fruits/fruits-360/test-multiple_fruits/Bananas(lady_finger)4.jpg")

img.show(figsize=(6,6))
```

These are some bananas, lets see if our model can predict on it.

```python
imgclass, y, idx = learn.predict(img)
```

```python
print(imgclass)
```
